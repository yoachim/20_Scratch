```python
df = pd.read_csv('types.dat', delimiter='|')
```

```python
def score_type(ptypes, df, types='All'):
    results = []
    if types== 'All':
        types = df['Type'].values
    for index, row in df.iterrows():
        temp = 0
        if row['Type'] in types:
            for ptype in ptypes:
                if ptype in row['StrongAgainst']:
                    temp -= 1
                if ptype in row['WeakAgainst']:
                    temp += 1
                if ptype in row['ResistantTo']:
                    temp -= 1
                if ptype in row['VulnerableTo']:
                    temp += 1
            results.append(temp)
    return results
```

```python
score_type(['Steel'], df, types=['Fairy'])
```
